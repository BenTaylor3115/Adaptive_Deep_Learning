# Perform time of all dataset

Now, you can select how many images you want to test. So first, select how many images (using the slider) you want to be performed and in which models the images will be performed in a Jetson Tx2 platform. This information will be collected to a database due to the time to collect allthe information can be too long.

```python
all_image_files = [f for f in listdir("images/val/images") if isfile(join("images/val/images", f))]


def f(amount_imgs):
    return random.sample(all_image_files, amount_imgs)

img_selected = interactive(f, amount_imgs=widgets.IntSlider(min=0,max=len(all_image_files),step=1,value=25000));
display(img_selected)

print("Models availables:")
models_selected_all_dataset = widgetFunctions.checkbox_list_creation(sorted(artefact.model_available_list()))
```

After the number of images performed are selected, the jetson platform will perform them. However, due to the huge time that it takes to perform more than 20 images, the information will be collected to our database where all the performance has been stored. Nevertheless, you can perform 20 images and the models selected.

```python
plotly.tools.set_credentials_file(username='visanz2', api_key='BmV5fZYP4BKMV4BKUun2')
data = []

# Name of the database
database_name = 'deep_learning'
list_correct_model_names, times_each_model_selected=database.get_best_models(database_name,img_selected.result, widgetFunctions.checkbox_models_selected(models_selected_all_dataset))


trace = go.Bar(
    x=list_correct_model_names,
    y=times_each_model_selected,
    name=""
)

data.append(trace)

layout = go.Layout(
    barmode='group',
    title='Times each model is selected as a best model using ' + str(len(img_selected.result)) + ' images',
    xaxis=dict(
        title='Models',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Times each model is selected ',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)


fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='grouped-bar')
    
```

The next code is showed the previous plot as a percetange of each model selected. You need to execute the previous code to obtain the plot of the following code.

```python
plotly.tools.set_credentials_file(username='visanz2', api_key='BmV5fZYP4BKMV4BKUun2')
data = []

percetange_model_selected = []

# This variable is in case that the slider has change and the previous code it wasn't performed
amount_images_performed = 0

for times_model in times_each_model_selected:
    amount_images_performed=amount_images_performed+times_model

if amount_images_performed > 0:
    for times_model in times_each_model_selected:
        percetange_model_selected.append(100*times_model/amount_images_performed)

trace = go.Bar(
    x=list_correct_model_names,
    y=percetange_model_selected,
    name=""
)

data.append(trace)

layout = go.Layout(
    barmode='group',
    title='How often a CNN model is considered to be optimal under the Top-1 score on ' + str(amount_images_performed) + ' images',
    xaxis=dict(
        title='Models',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='% of being optimal',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='grouped-bar')
    
```

# Premodel

First, you need to select how many images you want using our prototype (premodel selection and the model selected later)

```python
all_image_files = [f for f in listdir("images/val/images") if isfile(join("images/val/images", f))]

def f(amount_imgs):
    return amount_imgs 

img_features_selected = interactive(f, amount_imgs=widgets.IntSlider(min=0,max=len(all_image_files),step=1,value=25000));
display(img_features_selected)
```

Select which premodels you want to perform. The following machine learning models are used for the selection:

- Nearest Neighbour (nn)
- Decision Tree (dt16)
- Support Vector Machine (vc)

```python
premodel_selection = widgetFunctions.checkbox_list_creation(features.machines_avialable())
```

The following code shows the Top-1 accuracy of performing the amount of images selected of using different modeling techniques to form a 3-level premodel.

```python
list_premodels_selected = []

for element in widgetFunctions.checkbox_models_selected(premodel_selection):
    list_premodels_selected.append(re.findall(r"[\w']+", element))
    
percentage_results=features.prototype(img_features_selected.result,list_premodels_selected)

data = []

trace = go.Bar(
    x=widgetFunctions.checkbox_models_selected(premodel_selection),
    y=percentage_results,
    name=""
)

data.append(trace)

layout = go.Layout(
    barmode='group',
    title='Using different modeling techniques to form a 3-level premodel ',
    xaxis=dict(
        title='Premodels',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='% of success',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)


fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='grouped-bar')
    
```
